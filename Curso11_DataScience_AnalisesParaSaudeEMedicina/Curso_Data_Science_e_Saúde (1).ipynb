```
confirmed_by_country.loc["Mainland China"][2:]
```

```
import matplotlib.pyplot as plt

# grafico extremamente apertado
# estamos sendo longe do adequado ao passar uma mensagem errada

plt.figure(figsize=(2,5))
confirmed_by_country.loc["Mainland China"][2:].plot()
plt.title("Casos confirmados em 'Mainland China'")
plt.show()
```
